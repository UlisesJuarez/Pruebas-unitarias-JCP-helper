```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alumnos=pd.read_csv("alumnos.csv",encoding="ISO-8859-1",index_col=0)
```

```python
alumnos.head()
```

### Calificación por grupo

```python
grupo="3CM12"
datos_grupo=alumnos[alumnos["Grupo"]==grupo]
datos_grupo
```

```python
plt.figure(figsize=(8,5))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Id Alumno",fontsize=14)
plt.ylabel("Calificación",fontsize=14)
plt.title(f"Calificaciones del grupo: {grupo}")
plt.plot(datos_grupo.index,datos_grupo["Calificación"]);
```

##### Sí es más comodo para el docente podrá verlo en gráfico de barras

```python
fig, ax=plt.subplots(figsize=(10,6))
ax.barh(datos_grupo["Alumno"],datos_grupo["Calificación"])
ax.set(title=f"Calificaciones del grupo: {grupo}",
      xlabel="Calificación");
```

```python
alumnos.head()
```

<h3>Calificaciones por lenguaje de programación</h3>

```python
lenguaje="Python"
lenguajes=alumnos[alumnos["Lenguaje"]==lenguaje]
lenguajes
```

```python
prepara=lenguajes.groupby("Calificación").count()
```

```python
prepara.index
```

```python
prepara.Lenguaje.values
```

```python
plt.pie(prepara.Lenguaje.values, labels=prepara.index,autopct="%0.1f %%")
plt.axis("equal")
plt.title(f"Calificaciones de los alumnos en {lenguaje}")
plt.show()
```

<h5>Sí el docente lo prefiere podrá ver los datos en barras por ejemplo</h5>

```python
fig, ax=plt.subplots(figsize=(9,5))
ax.bar(prepara.index,prepara.Lenguaje.values)
ax.set(title=f"Calificaciones de los alumnos en {lenguaje}",
      xlabel="Calificación",ylabel="Número de alumnos");
```

```python
plt.figure(figsize=(8,5))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Id Alumno",fontsize=14)
plt.ylabel("Calificación",fontsize=14)
plt.title(f"Calificaciones del grupo: {grupo}")
plt.style.use("seaborn")
plt.scatter(datos_grupo.index,datos_grupo["Calificación"]);
```

```python
fig, ax=plt.subplots(figsize=(9,5))
ax.barh(datos_grupo["Alumno"],datos_grupo["Calificación"],color=['#005555','#069A8E','#417D7A','#035397','#066163'])
ax.set(title=f"Calificaciones del grupo: {grupo}",
      xlabel="Calificación");
```

```python
fig, ax=plt.subplots(figsize=(9,5))
ax.set_facecolor("#333")
ax.barh(datos_grupo["Alumno"],datos_grupo["Calificación"],color=['#B20600','#FF6363','#F32424','#FF4949','#FD5D5D',"#F56D91"])
ax.set(title=f"Calificaciones del grupo: {grupo}",
      xlabel="Calificación");
```

```python
colores = ['#BBBBBB','#7E7474',"#BDBDD7"]
```

```python
explode_vals = [0.1,0.1,0.1]
plt.figure(figsize=(8,4))
plt.pie(x=prepara.Lenguaje.values, labels=prepara.index,autopct="%0.1f %%",colors=colores,explode = explode_vals)
plt.axis("equal")
plt.title(f"Calificaciones de los alumnos en {lenguaje}")
plt.show()
```
